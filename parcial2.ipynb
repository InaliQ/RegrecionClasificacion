```python
import pandas as pd


columns_persona = ['idPersona', 'NombreCompleto', 'Domicilio', 'CodigoPostal', 'Nacionalidad', 'Telefono']
columns_medicamento = ['idMedicamento', 'Marca', 'NombreMedicamento']
columns_doctor = ['idDoctor', 'Cedula', 'NumeroDoctor', 'Horario', 'idPersona']
columns_enfermero = ['idEnfermero', 'Titulo', 'NumeroEnfermero', 'Horario', 'idPersona']
columns_padecimiento = ['idPadecimiento', 'Descripcion', 'Gravedad']
columns_paciente = ['idPaciente', 'NumeroPaciente', 'Altura', 'Peso', 'Sangre', 'idPersona']

df_persona = pd.read_csv('DimPersona.csv', header=None, names=columns_persona)
df_medicamento = pd.read_csv('DimMedicamento.csv', header=None, names=columns_medicamento)
df_doctor = pd.read_csv('DimDoctor.csv', header=None, names=columns_doctor)
df_enfermero = pd.read_csv('DimEnfermero.csv', header=None, names=columns_enfermero)
df_padecimiento = pd.read_csv('DimPadecimiento.csv', header=None, names=columns_padecimiento)
df_paciente = pd.read_csv('Paciente.csv', header=None, names=columns_paciente)

print("DimPersona")
display(df_persona.head())

print("DimMedicamento")
display(df_medicamento.head())

print("DimDoctor")
display(df_doctor.head())

print("DimEnfermero")
display(df_enfermero.head())

print("DimPadecimiento")
display(df_padecimiento.head())

print("Paciente")
display(df_paciente.head())


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df_persona, x='Nacionalidad', order=df_persona['Nacionalidad'].value_counts().index)
plt.title('Distribución de Nacionalidades')
plt.xticks(rotation=45)
plt.xlabel('Nacionalidad')
plt.ylabel('Número de Personas')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df_paciente, x='Altura', y='Peso', hue='Sangre', palette='viridis')
plt.title('Altura vs. Peso de los Pacientes')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')

plt.subplot(1, 2, 2)
sns.histplot(df_paciente['Altura'], kde=True, color='blue')
plt.title('Distribución de Altura de los Pacientes')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_paciente['Peso'], kde=True, color='green')
plt.title('Distribución del Peso de los Pacientes')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
plt.show()
```
